```
def build_batch_from_frame(frame, task, index):
    batch = {key: value for key, value in frame.items() if key.startswith("observation.")}
    batch["task"] = task
    batch["index"] = index
    return batch
```

```
prefix_att_2d_masks = make_att_2d_masks(prefix_pad_masks,
                                                prefix_att_masks)
prefix_position_ids = torch.cumsum(prefix_pad_masks, dim=1) - 1

prefix_att_2d_masks_4d = policy.model._prepare_attention_masks_4d(prefix_att_2d_masks)

policy.model.paligemma_with_expert.paligemma.language_model.config._attn_implementation = "eager"  # noqa: SLF001, E501

_, past_key_values = policy.model.paligemma_with_expert.forward(
            attention_mask=prefix_att_2d_masks_4d,
            position_ids=prefix_position_ids,
            past_key_values=None,
            inputs_embeds=[prefix_embs, None],
            use_cache=True,
        )
```

```
import torch
import torch.nn.functional as F
import gc

# --- Memory Cleanup ---
if 'past_key_values' in globals():
    del past_key_values
if 'prefill_hidden_states' in globals():
    del prefill_hidden_states
# Clear potential references in implicit variables
for var in ['_', '__', '___']:
    if var in globals():
        del globals()[var]
gc.collect()
torch.cuda.empty_cache()

# --- Configuration ---
max_decoding_steps = 100
temperature = 0.0
eos_token_id = tokenizer.eos_token_id

# Use inference_mode to disable gradient calculation and reduce memory usage
with torch.inference_mode():
    # --- 1. Prepare Initial State (Re-run prefill) ---
    prefill_hidden_states, past_key_values = policy.model.paligemma_with_expert.forward(
        attention_mask=prefix_att_2d_masks_4d,
        position_ids=prefix_position_ids,
        past_key_values=None,
        inputs_embeds=[prefix_embs, None],
        use_cache=True,
    )

    # Fix: prefill_hidden_states is a list [tensor, None], so we take the first element
    last_token_embedding = prefill_hidden_states[0][:, -1:]  # [B, 1, D]

    # Compute initial logits
    last_logits = policy.model.paligemma_with_expert.paligemma.lm_head(last_token_embedding)
    last_logits = F.log_softmax(last_logits, dim=-1).squeeze(1)

    # --- 2. Initialize Loop Variables ---
    batch_size = last_logits.shape[0]
    output_tokens = torch.zeros((batch_size, max_decoding_steps), dtype=torch.long, device=device)
    all_eos = torch.zeros(batch_size, dtype=torch.bool, device=device)
    prefill_len = prefix_pad_masks.shape[1]
    step = 0

    # --- 3. Generation Loop ---
    while step < max_decoding_steps and not all_eos.all():
        # A. Sample Token
        if temperature > 0.0:
            probs = torch.exp(last_logits / temperature)
            token = torch.multinomial(probs, num_samples=1)
        else:
            token = torch.argmax(last_logits, dim=-1, keepdim=True)  # Greedy

        # Store output
        output_tokens[:, step] = token.squeeze(-1)

        # Update EOS status
        all_eos |= (token.squeeze(-1) == eos_token_id)
        if all_eos.all():
            break

        # B. Prepare for next step
        token_embedding = policy.model.paligemma_with_expert.paligemma.language_model.embed_tokens(token)

        # Update position IDs
        current_position = prefill_len + step
        position_ids = torch.tensor([[current_position]] * batch_size, device=device, dtype=torch.long)

        # C. Decode one step using KV Cache
        decoder_out = policy.model.paligemma_with_expert.forward(
            inputs_embeds=[token_embedding, None],
            attention_mask=None,
            position_ids=position_ids,
            past_key_values=past_key_values,
            use_cache=True,
        )

        # Extract hidden states and update KV cache
        # Fix: decoder_out[0] is [tensor, None], so we take decoder_out[0][0]
        hidden_states = decoder_out[0][0]
        past_key_values = decoder_out[1]

        # Compute logits for next step
        last_logits = policy.model.paligemma_with_expert.paligemma.lm_head(hidden_states)
        last_logits = F.log_softmax(last_logits, dim=-1).squeeze(1)

        step += 1

# --- 4. Detokenize ---
print("Decoding output...")
decoded_texts = tokenizer.batch_decode(output_tokens, skip_special_tokens=True)
for i, text in enumerate(decoded_texts):
    print(f"Sample {i}: {text}")
```
